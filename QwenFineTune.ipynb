````python
class CustomerInfo(BaseModel):
    name: str = Field(..., description="Client name")
    phone: str = Field(..., description="Client number")
    email: str = Field(None, description="Client E-mail")
    account_or_order_id: Optional[str] = Field(None, description="account number or order number")


class IssueDetails(BaseModel):
    issue_type: str = Field(..., description="issue type")
    issue_description: str = Field(..., description="issue description")
    isSolved : bool = Field(..., description="is issue solved")




class CustomerServiceChecklist(BaseModel):
    Customer: CustomerInfo = Field(..., description="Customer information")
    Issue: IssueDetails = Field(..., description="Issue details")

````

````python
details_extraction_messages = [
    {
        "role": "system",
        "content": "\n".join([
            "You are an NLP data paraser.",
            "You will be provided by an Arabic for customer service call text associated with a Pydantic scheme.",
            "Generate the ouptut in the same script language.",
            "You have to extract JSON details from text according the Pydantic details.",
            "Extract details as mentioned in text.",
            "Do not generate any introduction or conclusion."
        ])
    },
    {
        "role": "user",
        "content": "\n".join([
            "## Chat:",
            chat1.strip(),
            "",

            "## Pydantic Details:",
            json.dumps(
                CustomerServiceChecklist.model_json_schema(), ensure_ascii=False
            ),
            "",

            "## Chat Details:",
            "```json"
        ])
    }
]
````

````python
import os
from openai import AzureOpenAI

endpoint = userdata.get("ENDPOINT_URL")
deployment = userdata.get("DEPLOYMENT_NAME")
subscription_key = userdata.get("AZURE_OPENAI_API_KEY")
````

````python
client = AzureOpenAI(
  azure_endpoint=endpoint,
  api_key=subscription_key,
  api_version="2025-01-01-preview",
)
````

````python
save_to = join(data_dir,  "conversation.jsonl")
````

````python
for i in range(200):

  if i%2 ==0:
    solv=" not solved"
  else:
    solv=" solved"
  prompt =f"""Generate conversation between customer and agent in call center
  Instruction:
  - the conversation must be in Egyptain Arabic.
  - the conversation must be between 150 to 300 words.
  - the conversation must include customer's name, phone.
  - the conversation may include customer's email, orderID.
  - the conversation must have an issue, the issue must be {solv}.
  - the conversation must not include any () like '(بعد دقيقة)'.
  - name of the customer must be in Arabic.
  - the customer may be male or female.



  Example:
  الموظف: أهلاً وسهلاً بحضرتك، معاك خدمة العملاء، إزاي أقدر أساعدك؟

العميل: عندي مشكله وعاوز حل ليها
  """
  chat_prompt = [
      {
        "role": "system",
        "content": prompt
      }
    ]

  completion = client.chat.completions.create(
          model=deployment,
          messages=chat_prompt,
          max_tokens=3000,
          temperature=0.9,
          top_p=0.9,
          frequency_penalty=0,
          presence_penalty=0,
          stop=None,
          stream=False,)
  with open(save_to,"a",encoding="utf-8") as f:
    f.write(json.dumps({"content":completion.choices[0].message.content},ensure_ascii=False, default=str)+"\n")
````

````python
save_to = join(data_dir,  "dataset.jsonl")
for i in range(len(dataset_list)):

  details_extraction_messages = [
      {
          "role": "system",
          "content": "\n".join([
              "You are an NLP data paraser.",
              "You will be provided by an Arabic for customer service call text associated with a Pydantic scheme.",
              "Generate the ouptut in the same script language.",
              "You have to extract JSON details from text according the Pydantic details.",
              "Extract details as mentioned in text.",
              "Do not generate any introduction or conclusion."
          ])
      },
      {
          "role": "user",
          "content": "\n".join([
              "## Chat:",
              dataset_list[i].strip(),
              "",

              "## Pydantic Details:",
              json.dumps(
                  CustomerServiceChecklist.model_json_schema(), ensure_ascii=False
              ),
              "",

              "## Chat Details:",
              "```json"
          ])
      }
  ]
  completion = client.chat.completions.create(
          model=deployment,
          messages=details_extraction_messages,
          max_tokens=3000,
          temperature=0.5,
          top_p=0.9,
          frequency_penalty=0,
          presence_penalty=0,
          stop=None,
          stream=False,)
  with open(save_to,"a",encoding="utf-8") as f:
    f.write(json.dumps({
        "id":i+1,
        "content":dataset_list[i],
        "task": "Extract the details of conversation into JSON.",
        "output_scheme": json.dumps(CustomerServiceChecklist.model_json_schema(),ensure_ascii=False),
        "response":completion.choices[0].message.content},
         ensure_ascii=False, default=str)+"\n")
````

## formating finetuning data

````python

sft_data_path = save_to
llm_finetunning_data = []

system_message = "\n".join([
    "You are a professional NLP data parser.",
    "Follow the provided `Task` by the user and the `Output Scheme` to generate the `Output JSON`.",
    "Do not generate any introduction or conclusion."
])

for line in open(sft_data_path):
    if line.strip() == "":
        continue

    rec = json.loads(line.strip())

    llm_finetunning_data.append({
        "system": system_message,
        "instruction": "\n".join([
            "# content:",
            rec["content"],

            "# Task:",
            rec["task"],

            "# Output Scheme:",
            rec["output_scheme"],
            "",

            "# Output JSON:",
            "```json"

        ]),
        "input": "",
        "output": "\n".join([
            "```json",
            json.dumps(rec["response"], ensure_ascii=False, default=str),
            "```"
        ]),
        "history": []
    })

random.Random(101).shuffle(llm_finetunning_data)
````

````python
llm_finetunning_data[0]
````

````python
train_sample_sz = 170

train_ds = llm_finetunning_data[:train_sample_sz]
eval_ds = llm_finetunning_data[train_sample_sz:]

os.makedirs(join(data_dir, "datasets", "llamafactory-finetune-data"), exist_ok=True)

with open(join(data_dir, "datasets", "llamafactory-finetune-data", "train.json"), "w") as dest:
    json.dump(train_ds, dest, ensure_ascii=False, default=str)

with open(join(data_dir, "datasets", "llamafactory-finetune-data", "val.json"), "w", encoding="utf8") as dest:
    json.dump(eval_ds, dest, ensure_ascii=False, default=str)
````
